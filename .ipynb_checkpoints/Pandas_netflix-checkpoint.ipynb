```python
#Quantos anos se passaram entre 2000 e o filme mais recente nos EUA
data[['lançamento']]\
.loc[(data['lançamento'] < 1999)\
     & (data['pais'] == 'United States')].max() \
- data[['lançamento']]\
.loc[(data['lançamento'] < 1999)\
     & (data['pais'] == 'United States')].min()
```

```python
#Quantos anos se passaram entre 1999 e o filme mais antigo nos EUA
data[['lançamento']]\
.loc[(data['lançamento'] < 2000)\
     & (data['pais'] == 'United States')].max() \
- data[['lançamento']]\
.loc[(data['lançamento'] < 2000)\
     & (data['pais'] == 'United States')].min()
```

```python
#Lançamentos entre 1999 e o filme mais antigo nos EUA
data[['titulo', 'lançamento', 'diretor', 'pais', 'adicionado']]\
.loc[(data['lançamento'] < 2000)\
     & (data['pais'] == 'United States')].count()
```

```python
#Lançamentos entre 2000 e o filme mais recente nos EUA
data[['titulo', 'lançamento', 'diretor', 'pais', 'adicionado']]\
.loc[(data['lançamento'] >= 2000)\
     & (data['pais'] == 'United States')].count()
```

```python
(data[['titulo', 'lançamento', 'diretor', 'pais', 'adicionado']]\
.loc[(data['lançamento'] >= 2000)\
     & (data['pais'] == 'United States')].count()\
/ data[['titulo', 'lançamento', 'diretor', 'pais', 'adicionado']]\
.loc[(data['lançamento'] < 2000)\
     & (data['pais'] == 'United States')].count()) * 100
```
